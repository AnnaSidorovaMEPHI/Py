```python
class HashMap:

    class Entry:
        def __init__(self, key, value=None):
            self.key=key
            self.value=value

        def get_key(self):
            return self.key

        def get_value(self):
            return self.value

        def __eq__(self, other):
            #TODO реализовать функцию сравнения
            #Пояснения: self - текущий объект entry, other - какой-то другой объект
            #Какое условие на равенство объектов в ассоциативном массиве?
            #Если объекты равны, то метод должен возвращать True, иначе False
            return(self.key == other.key)

    def __init__(self, bucket_num = 64):
        #TODO Создать массив для бакетов
        self.bucket_num=bucket_num#количество бакетов(пока пустых)
        self.size=0 #количество записей/Entries
        self.bucket=[None]*self.bucket_num#создали массив бакетов пустой массив на кол-во таких бакетов


    def get(self, key, default_value=None):
        #TODO метод get, возвращающий значение, если оно присутствует, иначе default_value
        i = self._get_index(self._get_hash(key))#нашли номер бакета по ключу через хэш
        it=self.Entry(key)#запомним
        if it in self.bucket[i]:#если объект есть в i-том бакете
            j=self.bucket[i].index(it)#индекс относительно этого бакета
            return self.bucket[i][j].get_value()
        else:
            return default_value
    def put(self, key, value):
        #TODO метод put, кладет значение по ключу, в случае, если ключ уже присутствует он его заменяет

        i = self._get_index(self._get_hash(key))#нашли номер бакета по ключу через хэш
        if self.bucket[i] is None:
            self.bucket[i]=[]
        it=self.Entry(key)#запомним
        if it in self.bucket[i]:#если объект есть в i-том бакете
            j=self.bucket[i].index(it)#индекс относительно этого бакета
            self.bucket[i][j].value=value#просто заменим
        else:
            self.bucket[i].append(self.Entry(key,value))#такого нет, добавим
            self.size+=1#кладем в бакет по ключу объект
    def __len__(self):
        #TODO Возвращает количество Entry в массиве/размер массива
        return self.size

    def _get_hash(self, key):
        #TODO Вернуть хеш от ключа, по которому он кладется в бакет
        return hash(key)
    def _get_index(self, hash_value):
        #TODO По значению хеша вернуть индекс элемента в массиве
        return hash_value%self.bucket_num

    def values(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор значений (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        for i in range(0,self.bucket_num):
            if self.bucket[i] is not None:
                for it in self.bucket[i]:
                    j=self.bucket[i].index(it)
                    yield self.bucket[i][j].value
    def keys(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор ключей (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        for i in range(0,self.bucket_num):
            if self.bucket[i] is not None:
                for it in self.bucket[i]:
                    j=self.bucket[i].index(it)
                    yield self.bucket[i][j].key
    def items(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор пар ключ и значение (tuples)
        #(на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        for i in range(0,self.bucket_num):
            if self.bucket[i] is not None:
                for it in self.bucket[i]:
                    j=self.bucket[i].index(it)
                    tuple=( self.bucket[i][j].key, self.bucket[i][j].value)
                    yield tuple
    def __str__(self):
        #TODO Метод, вызываемый при вызове print, пофантазируйте, что можно написать
        string=''
        for i in range(0,self.bucket_num):
            if self.bucket[i] is not None:
                for it in self.bucket[i]:
                    j=self.bucket[i].index(it)
                    string+= " { Key: "
                    string+=str(self.bucket[i][j].key)
                    string+= " ; Value:  "
                    string+=str(self.bucket[i][j].value)
                    string+="}"
                string+='\n'
        return(string)
```
